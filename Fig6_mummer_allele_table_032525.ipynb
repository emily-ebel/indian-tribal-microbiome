```python

def remove_items(test_list, item): 
    res = [i for i in test_list if i != item] 
    return res 
  
### Define a function that will generate a list of 'missing' positions -- i.e. those not covered by an alignment. 
def inverse_ranges(ranges, min_val, max_val):
    """Find the inverse of a list of ranges.
    Args:
        ranges (list of tuples): List of ranges, where each range is a tuple (start, end).
        min_val (int): The minimum value in the entire range.
        max_val (int): The maximum value in the entire range.
    Returns:
        list of tuples: The inverse of the input ranges.
    """
    inverse = []
    last_end = min_val - 1
    for start, end in sorted(ranges):
        if last_end < start - 1:
            inverse.append((last_end + 1, start - 1))
        last_end = end
    if last_end < max_val:
        inverse.append((last_end + 1, max_val))
    return inverse


def Intersection(lst1, lst2):
    return set(lst1).intersection(lst2)



def retain_longest_non_overlapping(segments):
    # Sort segments by their start point
    segments.sort(key=lambda x: x[0])
    
    longest_segments = []
    
    for current in segments:
        start, end = current
        length = end - start
        
        # Check if the current segment overlaps with any in the longest_segments
        overlaps = False
        for longest in longest_segments:
            l_start, l_end = longest
            # Check for overlap
            if not (end < l_start or start > l_end):
                overlaps = True
                # Retain the longest segment
                if length > (l_end - l_start):
                    longest_segments.remove(longest)
                    longest_segments.append(current)
                break
        
        # If no overlaps found, add the current segment
        if not overlaps:
            longest_segments.append(current)
    
    return longest_segments

# Example usage
#segments = [ (99335, 100000),(100165, 104015),(104418, 121389),(110316, 111341),  (111559, 112547), (121652, 122576), (123135, 153781)]
```

```python
###########################################################################
#######       WRITE TO FILE FOR PCA -- correct version as of 032625 
###########################################################################

# Header (position + samples)
# cells contain 1 for the rare allele, 0 for the common allele, or nothing if N

#outfile = open("/home/ebel/user_data/IndianMicrobiome/Bifidobacterium/adolescentis/mummer_files/ref_ATCC/Bif-adol_161_118975_PCA_102224.txt","w")

# in 111824 version, remove replicate genomes by editing SNV_table_FILT_102224.csv in R ...
# note this (R) fucks with the name,s but try anyway
outfile = open(path+"Bif-adol_157_113305_PCA_032625.txt","w")



#df_transposed.to_csv('/home/ebel/user_data/IndianMicrobiome/Bifidobacterium/adolescentis/mummer_files/ref_ATCC/SNV_table_102124.csv', index=True)
with open(path+"SNV_table_FILT_032526.csv","r") as csvfile:
#with open("/home/ebel/user_data/IndianMicrobiome/Bifidobacterium/adolescentis/mummer_files/ref_ATCC/SNV_table_FILT_102224.csv","r") as csvfile:
    reader = csv.reader(csvfile, delimiter=',') 
    header=next(reader)
    header[0] = 'position'

    for i in range(len(header)-1):
        outfile.write(header[i]+',')
    outfile.write(header[-1]+'\n')
        


    for row in reader:
        
        outfile.write(row[0]+',')
        data = row[1:len(row)]
        
        # replace N with blank
        data = ['' if x=='N' else x for x in data]     
        
        # identify the common and rare allele and replace with 0/1
        A_count = len([i for j,i in enumerate(data) if i =='A']) 
        T_count = len([i for j,i in enumerate(data) if i =='T'])
        C_count = len([i for j,i in enumerate(data) if i =='C']) 
        G_count = len([i for j,i in enumerate(data) if i =='G'])

        
        if max(A_count,T_count,C_count,G_count) == A_count: # A is the ref allele
            data = ['0' if x=='A' else x for x in data]
            data = ['1' if x in ['T','C','G'] else x for x in data]
            
        elif max(A_count,T_count,C_count,G_count) == T_count: # T is the ref allele
            data = ['0' if x=='T' else x for x in data]            
            data = ['1' if x in ['A','C','G'] else x for x in data]
            
        elif max(A_count,T_count,C_count,G_count) == C_count: # C is the ref allele
            data = ['0' if x=='C' else x for x in data]       
            data = ["1" if x in ['T','A','G'] else x for x in data]
        
        elif max(A_count,T_count,C_count,G_count) == G_count: # G is the ref allele
            data = ['0' if x=='G' else x for x in data]     
            data = ['1' if x in ['T','C','A'] else x for x in data]
        
        
        for i in range(len(data)-1):
            outfile.write(data[i]+',')
        outfile.write(data[-1]+'\n')
        


        

outfile.close()
```
